```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the train dataset
train_data = pd.read_csv('train.csv')
train_data.head()
```

```python
# Drop "Id" column since that will not be useful for machine learning.
train_data.drop('Id', axis=1, inplace=True)
train_data.head()
```

```python
# Determine the size of the dataframe
train_shape = train_data.shape
print(f"The train dataset has {train_shape[0]} instances and {train_shape[1]} features")
```

```python
# Determine the data types
train_data_types = train_data.dtypes
train_data_types, print(f"All data types are integers")
```

```python
# Describe the train dataset
pd.set_option('display.max_columns', None)
train_data.describe()
```

```python
# Determine the skewness of the distribution
train_data.skew()
```

```python
# Determine how many instances of Cover_Type there are
train_data.groupby('Cover_Type').size()
```

```python
# Since "Soil_Type7" and "Soil_Type15" are constants we can drop them from the dataframe.
train_data.drop(['Soil_Type7', 'Soil_Type15'], axis=1, inplace=True)
```

```python
# Verify columns are dropped.
train_data.head()
```

```python
# Save the cleaned dataframe as the new train dataset.
train_data.to_csv('clean_train.csv', index=False)
```

```python
# Load the test dataset
test_data = pd.read_csv('test.csv')
test_data.head()
```

```python
# Count the rows with Soil_Type7 and Soil_Type15 equal to 1
count_soil_type7 = len(test_data[test_data['Soil_Type7'] == 1])
count_soil_type15 = len(test_data[test_data['Soil_Type15'] == 1])

# Print the counts
print("Count of rows with Soil_Type7 = 1:", count_soil_type7)
print("Count of rows with Soil_Type15 = 1:", count_soil_type15)
```

```python
# Remove all rows in test.csv that have a value of 1 for "Soil_Type7" and "Soil_Type15"
test_data = test_data[(test_data['Soil_Type7'] != 1) & (test_data['Soil_Type15'] != 1)]
```

```python
# Verify the rows are gone.
count_soil_type7 = len(test_data[test_data['Soil_Type7'] == 1])
count_soil_type15 = len(test_data[test_data['Soil_Type15'] == 1])

# Print the counts
print("Count of rows with Soil_Type7 = 1:", count_soil_type7)
print("Count of rows with Soil_Type15 = 1:", count_soil_type15)
```

```python
# Remove the columns ""Soil_Type7" and "Soil_Type15".
test_data.drop(['Soil_Type7', 'Soil_Type15'], axis=1, inplace=True)
```
